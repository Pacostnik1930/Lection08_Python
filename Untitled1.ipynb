Задача № 6.Создать новый столбец в таблице с пингвинами,который будет отвечать за показатель длины келюва пингвина.high -длинный(от 42), middle - (от 35 до 42),low -маленький (до 35)

```python
def length(x):
  if x >= 42 :
    return 'high'
  elif x >= 35 and x <= 42:
    return 'middle'
  elif x <= 35:
    return 'low'     
penguins['length'] = penguins.bill_length_mm.apply(length)
```
